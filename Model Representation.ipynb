# Model Representation

```python
# import library
import numpy as np
import matplotlib.pyplot as plt
# plt.style.use('./deeplearning.mplstyle')
```

```python
# create data
x_train = np.array([1., 2.])
y_train = np.array([300., 500.])
print(f"x for train: {x_train}")
print(f"y for train: {y_train}")
```

```python
# explore data
print(f'shape of x_train: {x_train.shape}')
m = x_train.shape[0]
print(f'number of training examples: {m}')
```

```python
# plotting data
plt.scatter(x_train, y_train, marker='o', c='blue')
plt.title("House's Price")
plt.xlabel("Size of House")
plt.ylabel("Price of House")
```

```python
# Define model
weight = 100.
bias = 100.
print(f"Model:")
print(f"--> W: {Weight}")
print(f"--> b: {bias}")
```

```python
def predict_of_model(examples):
    predictions = np.zeros(m)
    for i in range(m):
        predictions[i] = weight * examples[i] + bias
    return predictions
```

```python
predictions = predict_of_model(x_train)
print(predictions)
```

```python
plt.scatter(x_train, y_train, marker='o', c='blue', label='Actual Value')

plt.plot(x_train, predictions, label='Predictions of Model')

plt.title("House's Price")
plt.xlabel("Size of House")
plt.ylabel("Price of House")
plt.legend()
```
