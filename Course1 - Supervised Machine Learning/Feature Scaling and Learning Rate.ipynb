## Normalization

```python
def z_score_normalization(x):
    
    mu = np.mean(x, axis=0)
    
    sigma = np.std(x, axis=0)
    
    x_norm = (x - mu) / sigma
    
    return (x_norm, mu, sigma)
```

```python
mu     = np.mean(x_train,axis=0)   
sigma  = np.std(x_train,axis=0) 
x_mean = (x_train - mu)
x_norm = (x_train - mu)/sigma 

fig, axes = plt.subplots(1, 3, figsize = (12, 3))

axes[0].scatter(x_train[:, 0], x_train[:, 3])
axes[0].set_title('unnormalizated')
axes[0].set_xlabel('size')
axes[0].set_ylabel('age')
axes[0].axis('equal')

axes[1].scatter(x_mean[:, 0], x_mean[:, 3])
axes[1].set_title('unnormalizated')
axes[1].set_xlabel('size')
axes[1].set_ylabel('age')
axes[1].axis('equal')

axes[2].scatter(x_norm[:, 0], x_norm[:, 3])
axes[2].set_title('unnormalizated')
axes[2].set_xlabel('size')
axes[2].set_ylabel('age')
axes[2].axis('equal')
```
