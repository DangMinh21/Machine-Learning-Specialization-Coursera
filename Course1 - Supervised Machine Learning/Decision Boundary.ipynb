```python
# packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
from lab_utils_common import sigmoid, plot_data, draw_vthresh
```

```python
# dataset
x = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1)
```

```python
fig, ax = plt.subplots(figsize=(4, 4))

pos = (y == 1).reshape(-1)
neg = (y == 0).reshape(-1)
print(pos)

ax.scatter(x[pos, 0], x[pos, 1], marker='x', c='red', label='positive')
ax.scatter(x[neg, 0], x[neg, 1], marker='o', c='blue', label='negative')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.legend()
plt.show()
```

```python
# model and plot decision boundary
fig, ax = plt.subplots(figsize=(4, 4))
x0 = np.arange(0, 3, 0.1)
x1 = 3 - x0

ax.scatter(x[pos, 0], x[pos, 1], marker='x', c='red', label='positive')
ax.scatter(x[neg, 0], x[neg, 1], marker='o', c='blue', label='negative')
ax.plot(x0, x1)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.legend()
plt.show()
```
