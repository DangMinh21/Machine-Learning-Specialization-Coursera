# Sigmoid or Logitic Regression

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# visualize sigmoid
x = np.arange(-7, 7, 0.1)
y = sigmoid(x)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Sigmoid Function')
plt.show()
```
